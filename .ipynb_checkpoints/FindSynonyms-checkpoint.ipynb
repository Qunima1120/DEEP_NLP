```python
import pandas as pd
```

```python
import spacy
```

```python
%%time
#nlp=spacy.load('en_vectors_web_lg')
nlp=spacy.load('./wikipedia-pubmed-and-PMC-w2v')
```

```python
embeddings = nlp.vocab.vectors.data
```

```python
#get the keys of the embedding table
myKeys=nlp.vocab.vectors.keys()
```

```python
from tqdm import tqdm
```

```python
keyToStrDict={}
failedCount=0
for key, _ in tqdm(nlp.vocab.vectors.items(),total=embeddings.shape[0]):
    try :
        keyToStrDict[key]=nlp.vocab.strings[key]
    except KeyError:
        continue
    
    #keyToStrDict[key]=nlp.vocab.strings[key]
    #print(key, nlp.vocab.strings[key], vector)
    #break
print ('{} failed'.format(failedCount))
```

```python
keyToStrS=pd.Series(keyToStrDict)
```

```python
type(key)
```

```python
for key, vector in nlp.vocab.vectors.items():
    print(key, nlp.vocab.strings[key], vector)
    break
```

```python
%time strKeyS=keyToStrS.reindex(myKeys)
```

```python
%time strCol=strKeyS.reindex(myKeys)
```

```python
embeddingDf=pd.DataFrame(index=strCol,data=embeddings)
```

```python
from scipy.spatial.distance import cosine
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
seedDf=embeddingDf.loc[['not available']]
```

```python
%time cosineSimM=cosine_similarity(seedDf,embeddingDf)
```

```python
cosineSimS=pd.Series(cosineSimM[0,:],index=embeddingDf.index)
```
